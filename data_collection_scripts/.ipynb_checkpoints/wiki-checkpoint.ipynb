```python
import requests 
from lxml import etree 
from bs4 import BeautifulSoup
import  pandas as pd
import numpy as np
```

```python
def under(s):
    return s.replace(' ','_')
```

```python
read_names_of_org = pd.read_csv('raw/all_org_name.csv', sep=',', header = None)
read_names_of_org[0] = read_names_of_org[0].apply(under)



```

```python
dictionary = {}
for names in read_names_of_org[0]:
    url = 'https://en.wikipedia.org/wiki/' + str(names)
    print(url)


    data = requests.get(url)  
    bs = BeautifulSoup(data.content, 'html.parser')
    
    title = bs.find('h1', class_ = 'firstHeading').text

    dictionary[title] = {}

    # print(dictionary)

    table = bs.find('table', class_ = 'infobox')
    if not table == None:
        table_body = table.find('tbody')
    # print(table_body)
#     text = table_body.get_text(strip=True,separator=',')
#     print(text)
        rows = table_body.find_all('tr')

        i = 0

        df = pd.DataFrame() #creates a new dataframe that's empty


        for row in rows:
    #         print(row.get_text(strip=True,separator=','))
            col = row.find('th', scope = "row")
            if not col == None:
                col_text = col.text
                value = row.find('td')
    #             print(col_text)
    #             print(value.get_text(strip=True,separator=','))
                dictionary[title][col_text] = value.get_text(strip=True,separator=',')

```

```python
dictionary
```

```python
df = pd.DataFrame.from_dict(dictionary)
```
