# WORKING SCRAPER CODE

```python
import requests 
from lxml import etree 
from bs4 import BeautifulSoup
import  pandas as pd
import numpy as np
import re
```

```python
df = pd.DataFrame()

urls = ['https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_January_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_February_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_March_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_April_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_May_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_June_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_July_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_August_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_September_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_October_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_November_2018',
       'https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_December_2018'
       ]

for url in urls:
    data = requests.get(url)  
    bs = BeautifulSoup(data.content, 'html.parser')

    table_data = bs.find('table', class_ = 'wikitable sortable')

    table_rows = table_data.find_all('tr')

    month_year = bs.find('h1', class_ = 'firstHeading')
    x = re.search("[ ][A-Za-z]{3,}[ ][0-9]{4}", month_year.text)


    res = []
    headers = []
    for tr in table_rows:
        th = table_data.find_all('th')
    #     print(th[0].text)
        header = [content.text.strip() for content in th if content.text.strip()]
        if header:
            header.append('month_year')
            headers.append(header)
        break

    for tr in table_rows:
    #     th = table_data.find_all('th')
        td = tr.find_all('td')
    #     print(th)
    #     print(td.text)
        row = [tr.text.strip() for tr in td if tr.text.strip()]
        if row:
            row.append(x.group())
            res.append(row)
    #         res.append(x.group())
    
    
    df2 = pd.DataFrame(res, columns=headers[0])
#     print(df)

    df2['Period'] = df2[['Date', 'month_year']].apply(lambda x: ''.join(x), axis=1)
    df2 = df2.drop(['Date', 'month_year'], axis=1)
    
    frames = [df, df2]
    df = pd.concat(frames)
```
