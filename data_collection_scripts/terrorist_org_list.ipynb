```python
import requests 
from lxml import etree 
from bs4 import BeautifulSoup
import  pandas as pd
import numpy as np
```

```python
url='https://en.wikipedia.org/wiki/List_of_designated_terrorist_groups'
```

```python

data = requests.get(url)  
bs = BeautifulSoup(data.content, 'html.parser')


# print(bs.prettify)
# print(dictionary)

table = bs.find('table', class_ = 'wikitable sortable')
# print(table)
table_body = table.find('tbody')
# # print(table_body)
rows = table_body.find_all('tr')
data = []
# print(rows)
for row in rows:

    cols = row.findAll('td')
    for col in cols:
        a = col.find('a')
        if not a == None:
            if a.text:
#                 print(a.text)
                data.append(a.text)
#         print(col)
# #     if not cols[1] == None:
# #         a = col[1].find('a')
# #         if not a == None:
# #              print(a.text)

#     print(type(cols))
```
